# Aqu√≠ se presentan los graficos como cientifico de datos

## Importar librerias

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = gpd.read_file('data/analisis_datos.gpkg')
data
```

```python
# Estad√≠sticas descriptivas b√°sicas
data.describe()
```

```python
single_step = data[data['time'] == 1800]
single_step
```

```python
# Histograma para la variable POBTOT (poblaci√≥n total)
plt.figure(figsize=(10, 6))
sns.histplot(single_step['POBTOT'], kde=True)
plt.title('Histograma de la poblaci√≥n total')
plt.xlabel('Poblaci√≥n total')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrama de caja para la variable Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Depth'])
plt.title('Diagrama de caja para la profundidad del agua')
plt.xlabel('Profundidad del agua')
plt.show()
```

```python
# Gr√°fico de densidad para la variable Rain (lluvia)
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Rain'], shade=True)
plt.title('Gr√°fico de densidad de la lluvia')
plt.xlabel('Lluvia')
plt.ylabel('Densidad')
plt.show()
```

```python
# Gr√°fico de dispersi√≥n entre la variable Velocity (velocidad del agua) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Velocity'], y=data['Depth'])
plt.title('Gr√°fico de dispersi√≥n entre la velocidad del agua y la profundidad del agua')
plt.xlabel('Velocidad del agua')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Mapa de calor para la correlaci√≥n entre las variables
correlation_matrix = data.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de calor de correlaci√≥n entre variables')
plt.show()
```

```python
# Gr√°fico de barras para la cantidad de escuelas por AGEB
barplot = data[data["time"] == 0]
barplot.reset_index(inplace=True)
plt.figure(figsize=(10, 6))
sns.barplot(x=barplot.index, y=barplot['escuela'])
plt.title('Cantidad de escuelas por AGEB')
plt.xlabel('√çndice AGEB')
plt.ylabel('Cantidad de escuelas')
plt.show()
```

```python
# Gr√°fico de barras agrupado para la cantidad de hospitales y farmacias por AGEB
ageb_index = barplot.index
plt.figure(figsize=(12, 6))
width = 0.4
plt.bar(ageb_index - width/2, barplot['hospital'], width, label='Hospitales')
plt.bar(ageb_index + width/2, barplot['farmacia'], width, label='Farmacias')
plt.title('Cantidad de hospitales y farmacias por AGEB')
plt.xlabel('√çndice AGEB')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

```python
# Gr√°fico de dispersi√≥n con ajuste de regresi√≥n lineal para la variable POBTOT (poblaci√≥n total) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.regplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gr√°fico de dispersi√≥n con ajuste de regresi√≥n lineal entre la poblaci√≥n total y la profundidad del agua')
plt.xlabel('Poblaci√≥n total')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Gr√°fico de dispersi√≥n con ajuste de regresi√≥n lineal para la variable POBTOT (poblaci√≥n total) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.regplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gr√°fico de dispersi√≥n con ajuste de regresi√≥n lineal entre la poblaci√≥n total y la profundidad del agua')
plt.xlabel('Poblaci√≥n total')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Gr√°fico de viol√≠n para la variable P30A59A_P (porcentaje de poblaci√≥n de 30 a 59 a√±os) agrupado por la cantidad de escuelas
plt.figure(figsize=(10, 6))
sns.violinplot(x=barplot['escuela'], y=barplot['P30A59A_P'], inner='quartile')
plt.title('Gr√°fico de viol√≠n del porcentaje de poblaci√≥n de 30 a 59 a√±os por cantidad de escuelas')
plt.xlabel('Cantidad de escuelas')
plt.ylabel('Porcentaje de poblaci√≥n de 30 a 59 a√±os')
plt.show()
```

```python
# Gr√°fico de pares (pairplot) para analizar relaciones entre un subconjunto de variables
subset_data = data[['Rain', 'Velocity', 'Depth', 'area', 'POBTOT']]
sns.pairplot(subset_data)
plt.show()
```

```python
# An√°lisis de valores faltantes en cada variable
missing_values = data.isnull().sum()
print("Valores faltantes por variable:\n", missing_values)
```

```python
# Calcular la matriz de correlaci√≥n de Pearson
correlation_matrix = data.corr()

# Imprimir la matriz de correlaci√≥n
print("Matriz de correlaci√≥n de Pearson:\n", correlation_matrix)

# Visualizar la matriz de correlaci√≥n con un mapa de calor (heatmap)
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de la matriz de correlaci√≥n de Pearson')
plt.show()

# Calcular la matriz de correlaci√≥n de Spearman
spearman_correlation_matrix = data.corr(method='spearman')

# Visualizar la matriz de correlaci√≥n de Spearman con un mapa de calor
plt.figure(figsize=(14, 10))
sns.heatmap(spearman_correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de la matriz de correlaci√≥n de Spearman')
plt.show()

# Gr√°fico de dispersi√≥n para analizar la correlaci√≥n entre dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gr√°fico de dispersi√≥n entre la poblaci√≥n total y la profundidad del agua')
plt.xlabel('Poblaci√≥n total')
plt.ylabel('Profundidad del agua')
plt.show()
```
