# Aquí se presentan los graficos como cientifico de datos

## Importar librerias

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = gpd.read_file('data/analisis_datos.gpkg')
data
```

```python
# Estadísticas descriptivas básicas
data.describe()
```

```python
single_step = data[data['time'] == 1800]
single_step
```

```python
# Histograma para la variable POBTOT (población total)
plt.figure(figsize=(10, 6))
sns.histplot(single_step['POBTOT'], kde=True)
plt.title('Histograma de la población total')
plt.xlabel('Población total')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrama de caja para la variable Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Depth'])
plt.title('Diagrama de caja para la profundidad del agua')
plt.xlabel('Profundidad del agua')
plt.show()
```

```python
# Gráfico de densidad para la variable Rain (lluvia)
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Rain'], shade=True)
plt.title('Gráfico de densidad de la lluvia')
plt.xlabel('Lluvia')
plt.ylabel('Densidad')
plt.show()
```

```python
# Gráfico de dispersión entre la variable Velocity (velocidad del agua) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Velocity'], y=data['Depth'])
plt.title('Gráfico de dispersión entre la velocidad del agua y la profundidad del agua')
plt.xlabel('Velocidad del agua')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Mapa de calor para la correlación entre las variables
correlation_matrix = data.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de calor de correlación entre variables')
plt.show()
```

```python
# Gráfico de barras para la cantidad de escuelas por AGEB
barplot = data[data["time"] == 0]
barplot.reset_index(inplace=True)
plt.figure(figsize=(10, 6))
sns.barplot(x=barplot.index, y=barplot['escuela'])
plt.title('Cantidad de escuelas por AGEB')
plt.xlabel('Índice AGEB')
plt.ylabel('Cantidad de escuelas')
plt.show()
```

```python
# Gráfico de barras agrupado para la cantidad de hospitales y farmacias por AGEB
ageb_index = barplot.index
plt.figure(figsize=(12, 6))
width = 0.4
plt.bar(ageb_index - width/2, barplot['hospital'], width, label='Hospitales')
plt.bar(ageb_index + width/2, barplot['farmacia'], width, label='Farmacias')
plt.title('Cantidad de hospitales y farmacias por AGEB')
plt.xlabel('Índice AGEB')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

```python
# Gráfico de dispersión con ajuste de regresión lineal para la variable POBTOT (población total) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.regplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gráfico de dispersión con ajuste de regresión lineal entre la población total y la profundidad del agua')
plt.xlabel('Población total')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Gráfico de dispersión con ajuste de regresión lineal para la variable POBTOT (población total) y Depth (profundidad del agua)
plt.figure(figsize=(10, 6))
sns.regplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gráfico de dispersión con ajuste de regresión lineal entre la población total y la profundidad del agua')
plt.xlabel('Población total')
plt.ylabel('Profundidad del agua')
plt.show()
```

```python
# Gráfico de violín para la variable P30A59A_P (porcentaje de población de 30 a 59 años) agrupado por la cantidad de escuelas
plt.figure(figsize=(10, 6))
sns.violinplot(x=barplot['escuela'], y=barplot['P30A59A_P'], inner='quartile')
plt.title('Gráfico de violín del porcentaje de población de 30 a 59 años por cantidad de escuelas')
plt.xlabel('Cantidad de escuelas')
plt.ylabel('Porcentaje de población de 30 a 59 años')
plt.show()
```

```python
# Gráfico de pares (pairplot) para analizar relaciones entre un subconjunto de variables
subset_data = data[['Rain', 'Velocity', 'Depth', 'area', 'POBTOT']]
sns.pairplot(subset_data)
plt.show()
```

```python
# Análisis de valores faltantes en cada variable
missing_values = data.isnull().sum()
print("Valores faltantes por variable:\n", missing_values)
```

```python
# Calcular la matriz de correlación de Pearson
correlation_matrix = data.corr()

# Imprimir la matriz de correlación
print("Matriz de correlación de Pearson:\n", correlation_matrix)

# Visualizar la matriz de correlación con un mapa de calor (heatmap)
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de la matriz de correlación de Pearson')
plt.show()

# Calcular la matriz de correlación de Spearman
spearman_correlation_matrix = data.corr(method='spearman')

# Visualizar la matriz de correlación de Spearman con un mapa de calor
plt.figure(figsize=(14, 10))
sns.heatmap(spearman_correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de la matriz de correlación de Spearman')
plt.show()

# Gráfico de dispersión para analizar la correlación entre dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['POBTOT'], y=data['Depth'])
plt.title('Gráfico de dispersión entre la población total y la profundidad del agua')
plt.xlabel('Población total')
plt.ylabel('Profundidad del agua')
plt.show()
```
